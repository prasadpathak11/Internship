```python
# Question- 1 (Wikipedia)

# Fetch the HTML content from Wikipedia.org
url = "https://en.wikipedia.org/wiki/Main_Page"
response = requests.get(url)
html_content = response.content

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Find all header tags (h1 to h6)
header_tags = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])

# Extract the text from header tags
header_texts = [header.get_text() for header in header_tags]

# Create a dataframe
df = pd.DataFrame({"Headers": header_texts})

# Display the dataframe
print(df)
```
